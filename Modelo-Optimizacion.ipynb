# Lectura de datos(Todas las veces menos la primera vez que los extraje)


```python


# 📁 Carpeta donde están los archivos en Drive
carpeta = "/content/drive/MyDrive/TFM/TFM"

# 🔍 Buscamos todos los xlsx que empiecen por "precios_"
patron = os.path.join(carpeta, "precios_*.xlsx")
archivos = glob.glob(patron)

print("Archivos encontrados:")
for a in archivos:
    print(" -", os.path.basename(a))

# 📥 Leemos cada archivo y guardamos su DataFrame
dataframes = []

for archivo in archivos:
    df = pd.read_excel(archivo)

    # Localizamos la columna de fechas (se llame "date", "Date", etc.)
    col_fecha = None
    for col in df.columns:
        if str(col).strip().lower() == "date":
            col_fecha = col
            break

    if col_fecha is None:
        print(f"⚠️  {os.path.basename(archivo)} no tiene columna 'date', se omite.")
        continue

    # Convertimos a fecha y la ponemos como índice
    df[col_fecha] = pd.to_datetime(df[col_fecha])
    df = df.set_index(col_fecha)
    df.index.name = "Date"

    dataframes.append(df)

# 🔗 Unimos todos los DataFrames por fecha (índice común)
# axis=1 -> cada activo es una columna; se alinean por fecha
prices_raw = pd.concat(dataframes, axis=1)

# Ordenamos por fecha
prices_raw = prices_raw.sort_index()
```

# Limpieza de datos

```python
# 🧹 limpiar NaN iniciales (algunos activos empiezan más tarde)
prices = prices_raw.dropna()
```

```python
#Me quedo con años de 2021 a 2025 ambos incluidos
prices=prices.loc["2021-01-01":"2025-12-31"]
```

```python
prices.rename(columns={"SP500": "SPY","EuroStoxx50":"FEZ","Nikkei225":"EWJ","CSI300":"ASHR","MSCI EmergingMarkets":"EEM","Australia":"EWA","FTSE100":"EWU","World":"URTH","Nasdaq100":"QQQ","AI_BigData":"AIQ",
                       "CrudeOil":"USO","naturalGas":"UNG","gold":"GLD","Silver":"SLV","Cobre":"CPER","Platinum":"JJU","Soja":"SOYB","Trigo":"WEAT","Madera":"WOOD","Corn":"CORN"}, inplace=True)
```

```python

#Rentabilidad logaritmica
rentabilidad_log = np.log(prices / prices.shift(1))
# Eliminamos la primera fila ya que el primer día no tiene día anterior por lo que no se vera la rentabilidad
rentabilidad_log = rentabilidad_log.dropna()
```

## Rendimiento esperado

```python
# Rendimiento esperado
rendimiento_esperado = rentabilidad_log.mean()
# Tabla
tabla_re = rendimiento_esperado.reset_index()
tabla_re.columns = ["Activo", "E(Ri)"]
tabla_re = tabla_re.sort_values("E(Ri)", ascending=False).reset_index(drop=True)
```

# Rendimiento ajustado al riesgo(Ratio de Sharpe)

```python
rf_anual = 0.02#Letras del Tesoro español a 12 meses entre 2021 y 2025(Figura 4.9)
rf_diaria = np.log(1 + rf_anual) / 252

# Rentabilidad acumulada del período (no esperanza)
rp = rentabilidad_log.sum()  # suma de log-retornos = rentabilidad log acumulada
rf_periodo = rf_diaria * len(rentabilidad_log)  # Rf escalada al mismo período

sharpe = (rp - rf_periodo) / rentabilidad_log.std()

tabla_sharpe = sharpe.reset_index()
tabla_sharpe.columns = ["Activo", "Sharpe Ratio"]
tabla_sharpe = tabla_sharpe.sort_values("Sharpe Ratio", ascending=False)
print(tabla_sharpe.to_string(index=False))

```

```python
fig, ax = plt.subplots(figsize=(12, 8))

colors = ["#2ecc71" if v > 0 else "#e74c3c" for v in tabla_sharpe["Sharpe Ratio"]]

ax.hlines(y=tabla_sharpe["Activo"], xmin=0, xmax=tabla_sharpe["Sharpe Ratio"],
          color=colors, linewidth=2, alpha=0.7)
ax.scatter(tabla_sharpe["Sharpe Ratio"], tabla_sharpe["Activo"],
           color=colors, s=20, zorder=2)

for _, row in tabla_sharpe.iterrows():
    offset = 0.3 if row["Sharpe Ratio"] >= 0 else -0.3
    ax.text(row["Sharpe Ratio"] + offset, row["Activo"],
            f"{row['Sharpe Ratio']:.3f}", va="center",
            ha="left" if row["Sharpe Ratio"] >= 0 else "right",
            fontsize=10, fontweight="bold")

ax.axvline(0, color="gray", linewidth=0.8, linestyle="--", alpha=0.6)
ax.set_title("Sharpe Ratio por Activo", fontsize=16, pad=15)

# Ajustar estos valores según el rango real
x_min = tabla_sharpe["Sharpe Ratio"].min()
x_max = tabla_sharpe["Sharpe Ratio"].max()
ax.set_xlim(x_min - 8, x_max + 8)

plt.tight_layout()
plt.show()
```

#4.5 Costes de transacción, vetnana, rebalanceo, metodo de contraste

#Costes de transacción

```python
MIN_MESES_TEST         = 12       # mínimo de periodos fuera de muestra exigido a cualquier configuración
COSTE_BPS              = 10       # coste unitario kappa, en puntos básicos sobre el volumen negociado
CAPITAL_INICIAL        = 100.0    # C0: aportación inicial (base 100; resultados invariantes a la escala)
APORTACION_MENSUAL_PCT = 0.1     # aportación mensual como proporción de C0 (constante en euros)

meses_periodo   = rentabilidad_log.index.to_period("M")
meses_unicos    = meses_periodo.unique().sort_values()
n_meses_totales = len(meses_unicos)


def pesos_markowitz(returns_window, rf=rf_anual, weight_bounds=(0, 1)):
    """
    Pesos de la cartera tangente (máximo Ratio de Sharpe) sobre la ventana dada.
    La matriz de covarianzas se regulariza con contracción de Ledoit-Wolf. Si
    ningún activo bate a la tasa libre de riesgo en la ventana (max_sharpe no
    tiene solución long-only), se recurre a la cartera de mínima varianza.
    """
    mu = returns_window.mean() * 252
    S  = risk_models.CovarianceShrinkage(returns_window, returns_data=True, frequency=252).ledoit_wolf()
    ultimo_error = None

    for solver in [None, "ECOS", "SCS"]:       # Intento 1: máximo Ratio de Sharpe
        try:
            ef = (EfficientFrontier(mu, S, weight_bounds=weight_bounds, solver=solver)
                  if solver else EfficientFrontier(mu, S, weight_bounds=weight_bounds))
            ef.max_sharpe(risk_free_rate=rf)
            pesos = pd.Series(ef.clean_weights()); pesos = pesos[pesos > 0]
            return pesos / pesos.sum()
        except Exception as e:
            ultimo_error = e; continue

    for solver in [None, "ECOS", "SCS"]:       # Intento 2 (respaldo): mínima varianza
        try:
            ef = (EfficientFrontier(mu, S, weight_bounds=weight_bounds, solver=solver)
                  if solver else EfficientFrontier(mu, S, weight_bounds=weight_bounds))
            ef.min_volatility()
            pesos = pd.Series(ef.clean_weights()); pesos = pesos[pesos > 0]
            return pesos / pesos.sum()
        except Exception as e:
            ultimo_error = e; continue

    raise RuntimeError(f"Ningún solver pudo resolver la optimización: {ultimo_error}")


def coste_rebalanceo_con_aportacion(pesos_previos, pesos_objetivo, universo,
                                    valor_previo, aportacion, coste_bps=COSTE_BPS):
    """
    Coste de un rebalanceo con aportación periódica, resuelto en DOS NIVELES.

    'aportacion' (A_t) es el importe de capital nuevo recibido en este rebalanceo,
    en las mismas unidades que 'valor_previo'. Bajo el esquema adoptado, ese
    importe es constante en el tiempo e igual a la aportación mensual (un
    porcentaje del capital inicial C0) multiplicada por el número de meses
    transcurridos desde el rebalanceo anterior:  A_t = a * C0 * m_t.

    1) El capital total disponible pasa a ser  V_t = V_t^- + A_t.
    2) Para cada activo:  Delta_i = w*_i * V_t - v_i^-   (>0 comprar, <0 vender).
       La aportación cubre primero las compras de los infraponderados; solo el
       exceso de los sobreponderados obliga a vender posiciones existentes.
    Identidad de conservación (los pesos suman 1):
       necesidad_compra - necesidad_venta = A_t.
    El coste se cobra sobre TODO el volumen efectivamente negociado.
    """
    valor_total = valor_previo + aportacion

    w_previo   = pd.Series(0.0, index=universo); w_previo.update(pesos_previos)
    w_objetivo = pd.Series(0.0, index=universo); w_objetivo.update(pesos_objetivo)

    valor_actual_activo   = w_previo   * valor_previo     # v_i^-  (antes de aportar)
    valor_objetivo_activo = w_objetivo * valor_total      # w*_i * V_t (tras aportar)
    delta = valor_objetivo_activo - valor_actual_activo

    necesidad_compra = delta[delta > 0].sum()
    necesidad_venta  = (-delta[delta < 0]).sum()

    # Descomposición del volumen negociado:
    compra_con_aportacion = min(necesidad_compra, aportacion)                          # despliegue del dinero nuevo
    compraventa_existente = necesidad_venta + max(necesidad_compra - aportacion, 0.0)  # trading de posiciones

    volumen_negociado = compra_con_aportacion + compraventa_existente
    turnover = volumen_negociado / valor_total
    coste    = turnover * (coste_bps / 10000)

    return {"turnover": turnover, "coste_%": coste, "valor_total": valor_total,
            "aportacion": aportacion,
            "turnover_aportacion": compra_con_aportacion / valor_total,
            "turnover_posiciones": compraventa_existente / valor_total}
```

#  Procedimiento de contraste

```python
#Tabla 4.1
def ejecutar_backtest(rentabilidad_log, lookback, freq, window_type="rolling",
                      coste_bps=COSTE_BPS, aportacion_mensual_pct=APORTACION_MENSUAL_PCT,
                      capital_inicial=CAPITAL_INICIAL, rf=rf_anual,
                      min_test_months=MIN_MESES_TEST, benchmark="SPY"):
    universo   = list(rentabilidad_log.columns)
    primer_idx = lookback
    indices_rebalanceo = [i for i in range(primer_idx, n_meses_totales, freq)]
    meses_test = n_meses_totales - primer_idx
    if meses_test < min_test_months:
        raise ValueError(f"Quedarían solo {meses_test} meses de test (mínimo exigido: {min_test_months}).")

    # Aportación mensual CONSTANTE (porcentaje) x (capital inicial C0).
    aportacion_mensual = aportacion_mensual_pct * capital_inicial

    valor = capital_inicial            # C0: aportación inicial que constituye la cartera
    curva_partes, log_rebalanceos, retornos_mensuales = [], [], []
    pesos_historicos = {}
    pesos_actuales = None

    for k, i_reb in enumerate(indices_rebalanceo):
        meses_train  = (meses_unicos[0:i_reb] if window_type == "expanding"
                        else meses_unicos[i_reb - lookback:i_reb])
        fechas_train = rentabilidad_log.index[meses_periodo.isin(meses_train)]
        pesos_nuevos = pesos_markowitz(rentabilidad_log.loc[fechas_train], rf=rf)
        fecha_reb    = rentabilidad_log.index[meses_periodo == meses_unicos[i_reb]][0]
        pesos_historicos[fecha_reb] = pesos_nuevos

        if pesos_actuales is None:
            # Constitución de la cartera con el capital inicial C0 (aportación inicial).
            reb = {"fecha": fecha_reb, "turnover": 1.0, "coste_%": coste_bps / 10000,
                   "aportacion": 0.0, "turnover_aportacion": 0.0, "turnover_posiciones": 0.0}
            valor *= (1 - reb["coste_%"])
        else:
            meses_transcurridos = i_reb - indices_rebalanceo[k - 1]     # = freq entre rebalanceos
            aportacion = aportacion_mensual * meses_transcurridos        # A_t = a * C0 * m_t
            info  = coste_rebalanceo_con_aportacion(
                pesos_actuales, pesos_nuevos, universo, valor, aportacion, coste_bps)
            valor = info["valor_total"] * (1 - info["coste_%"])
            reb   = {"fecha": fecha_reb, **info}
        log_rebalanceos.append(reb)

        i_fin       = indices_rebalanceo[k + 1] if k + 1 < len(indices_rebalanceo) else n_meses_totales
        meses_hold  = meses_unicos[i_reb:i_fin]
        fechas_hold = rentabilidad_log.index[meses_periodo.isin(meses_hold)]
        sub         = rentabilidad_log.loc[fechas_hold, pesos_nuevos.index]
        rel_precio  = np.exp(sub.cumsum())
        valor_diario = valor * rel_precio.dot(pesos_nuevos)
        curva_partes.append(valor_diario)

        # Rentabilidad mensual de inversión del periodo (sin computar las aportaciones).
        # Se ancla el valor inicial en el ÚLTIMO día del mes ANTERIOR, para que el primer
        # mes del periodo siempre tenga base de cálculo aunque no empiece a negociar el día 1.
        mes_prev = (fechas_hold[0].to_period("M") - 1).to_timestamp("M")
        vm       = pd.concat([pd.Series([valor], index=[mes_prev]), valor_diario])
        vm       = vm.groupby(vm.index.to_period("M")).last()
        retornos_mensuales.append(vm.pct_change().dropna())
        pesos_actuales  = pesos_nuevos * rel_precio.iloc[-1]
        pesos_actuales /= pesos_actuales.sum()
        valor = valor_diario.iloc[-1]

    curva              = pd.concat(curva_partes)
    retornos_mensuales = pd.concat(retornos_mensuales)
    rent_anual = retornos_mensuales.mean() * 12
    vol_anual  = retornos_mensuales.std() * np.sqrt(12)

    return {
        "curva": curva,
        "retornos_mensuales": retornos_mensuales,
        "log_rebalanceos": pd.DataFrame(log_rebalanceos),
        "pesos_historicos": pesos_historicos,
        "metricas": {
            "Rentabilidad anual neta": rent_anual,
            "Volatilidad anual": vol_anual,
            "Sharpe neto": (rent_anual - rf) / vol_anual,
            "Coste total acumulado": 1 - np.prod([1 - r["coste_%"] for r in log_rebalanceos]),
            "Nº rebalanceos": len(indices_rebalanceo),
            "Meses de test": meses_test,
        },
    }


resultados        = {nombre: ejecutar_backtest(rentabilidad_log, **p) for nombre, p in modelos.items()}
tabla_comparativa = pd.DataFrame({n: r["metricas"] for n, r in resultados.items()}).T
tabla_comparativa
```

```python
#Tabla 4.4
def jobson_korkie_memmel(ret_i, ret_j, rf_anual=rf_anual, freq=12):
    """
    Contraste de igualdad de ratios de Sharpe (Jobson y Korkie, 1981; corrección
    de Memmel, 2003). Trabaja con retornos EXCESO de la misma frecuencia y
    devuelve el estadístico z ~ N(0,1) y su p-valor bilateral. H0: SR_i = SR_j.
    """
    df = pd.concat([ret_i, ret_j], axis=1, join="inner").dropna()
    rf_periodo = rf_anual / freq
    e_i = df.iloc[:, 0] - rf_periodo
    e_j = df.iloc[:, 1] - rf_periodo
    T   = len(df)

    sr_i = e_i.mean() / e_i.std(ddof=1)
    sr_j = e_j.mean() / e_j.std(ddof=1)
    rho  = e_i.corr(e_j)

    theta = (1.0 / T) * (2 * (1 - rho)
                         + 0.5 * (sr_i**2 + sr_j**2 - 2 * sr_i * sr_j * rho**2))
    z = (sr_i - sr_j) / np.sqrt(theta)
    p = 2 * (1 - stats.norm.cdf(abs(z)))

    escala = np.sqrt(freq)
    return {"SR_i_anual": sr_i * escala, "SR_j_anual": sr_j * escala,
            "z": z, "p_valor": p, "T": T, "rho": rho}


# ------------------------------------------------------------------ #
# Contraste de TODOS contra TODOS (matriz por parejas)
# ------------------------------------------------------------------ #
nombres = list(resultados.keys())
n = len(nombres)

# Matrices de estadístico z, p-valor y nº de meses comunes de cada pareja.
# Convención: la celda (fila i, columna j) contrasta SR_i frente a SR_j.
#   z > 0  -> el modelo de la FILA tiene mayor Sharpe que el de la COLUMNA.
#   z es antisimétrica (z_ij = -z_ji); el p-valor y los meses comunes son simétricos.
Z = pd.DataFrame(np.nan, index=nombres, columns=nombres)
P = pd.DataFrame(np.nan, index=nombres, columns=nombres)
Tcom = pd.DataFrame(np.nan, index=nombres, columns=nombres)

for a in range(n):
    for b in range(n):
        if a == b:
            continue
        out = jobson_korkie_memmel(resultados[nombres[a]]["retornos_mensuales"],
                                   resultados[nombres[b]]["retornos_mensuales"],
                                   rf_anual=rf_anual, freq=12)
        Z.iloc[a, b]    = round(out["z"], 3)
        P.iloc[a, b]    = round(out["p_valor"], 4)
        Tcom.iloc[a, b] = out["T"]

# Sharpe anualizado de cada modelo (sobre su propio periodo de test), como referencia.
sharpe_modelos = pd.Series(
    {nombre: r["metricas"]["Sharpe neto"] for nombre, r in resultados.items()},
    name="Sharpe neto"
).round(3)

# Resumen de parejas estadísticamente significativas al 95 % (p < 0.05).
ALPHA = 0.05
significativas = []
for a in range(n):
    for b in range(a + 1, n):          # solo mitad superior: cada pareja una vez
        p_ab = P.iloc[a, b]
        if pd.notna(p_ab) and p_ab < ALPHA:
            z_ab = Z.iloc[a, b]
            mejor = nombres[a] if z_ab > 0 else nombres[b]
            significativas.append({
                "Modelo i": nombres[a], "Modelo j": nombres[b],
                "z (i vs j)": z_ab, "p-valor": p_ab,
                "Meses comunes": int(Tcom.iloc[a, b]),
                "Mayor Sharpe": mejor,
            })
significativas = pd.DataFrame(significativas)

print("Sharpe neto anualizado por modelo (periodo de test propio):")
display(sharpe_modelos.to_frame())

print("\nMatriz de estadísticos z (fila vs columna; z>0 -> la fila tiene mayor Sharpe):")
display(Z)

print("\nMatriz de p-valores (bilateral, simétrica):")
display(P)

print(f"\nParejas significativas al {int((1-ALPHA)*100)}% de confianza (p < {ALPHA}):")
if significativas.empty:
    print("  Ninguna: a este nivel de confianza no se rechaza la igualdad de "
          "ratios de Sharpe en ninguna pareja.")
else:
    display(significativas.set_index(["Modelo i", "Modelo j"]))
```

## Contraste estadístico frente al S&P 500


```python
#Tabla 5.2
def jobson_korkie_memmel(ri, rj, rf_anual=rf_anual, freq=12):
    """Jobson-Korkie (1981) con corrección de Memmel (2003). H0: SR_i = SR_j."""
    df = pd.concat([ri, rj], axis=1, join="inner").dropna()
    rf_p = rf_anual/freq
    ei, ej = df.iloc[:,0]-rf_p, df.iloc[:,1]-rf_p
    T = len(df)
    si, sj = ei.mean()/ei.std(ddof=1), ej.mean()/ej.std(ddof=1)
    rho = ei.corr(ej)
    theta = (1/T)*(2*(1-rho) + 0.5*(si**2 + sj**2 - 2*si*sj*rho**2))
    z = (si - sj)/np.sqrt(theta); p = 2*(1-stats.norm.cdf(abs(z)))
    return {"SR_H": si*np.sqrt(freq), "SR_SP": sj*np.sqrt(freq), "rho": rho, "T": T, "z": z, "p": p}

ALPHA = 0.05
jk = jobson_korkie_memmel(ret_h, ret_spy)
t_stat, p_t = stats.ttest_rel(ret_h, ret_spy)

print(f"Meses comunes: {jk['T']}    correlación H-S&P: {jk['rho']:.3f}")
print(f"Sharpe anual   ->  Modelo H: {jk['SR_H']:.3f}    S&P 500: {jk['SR_SP']:.3f}\n")
print("Jobson-Korkie-Memmel (diferencia de ratios de Sharpe):")
print(f"   z = {jk['z']:.3f}   p = {jk['p']:.4f}   -> "
      f"{'H mejor, SIGNIFICATIVO al 95%' if jk['p']<ALPHA and jk['z']>0 else 'NO significativo al 95%'}\n")
print("t apareado (diferencia de rentabilidad mensual media):")
print(f"   t = {t_stat:.3f}   p = {p_t:.4f}   -> "
      f"{'SIGNIFICATIVO al 95%' if p_t<ALPHA else 'NO significativo al 95%'}\n")
if (jk['p']<ALPHA and jk['z']>0) or p_t<ALPHA:
    print("Conclusión: hay evidencia al 95% de que el Modelo H supera al S&P 500.")
else:
    print("Conclusión: las estimaciones puntuales favorecen al Modelo H, pero con este\n"
          "periodo la diferencia NO es estadísticamente significativa al 95%.")
```
